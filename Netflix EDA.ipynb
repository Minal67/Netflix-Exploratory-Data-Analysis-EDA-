## <FONT COLOR='Aqua'>EXPLORATORY ANALYSIS AND VISUALIZATION

```python
#sorting according to rating
netflix_data.sort_values("rating")
```

```python
#sorting according to rating highest to lowest
netflix_data.sort_values("rating",ascending=False)
```

```python
#top 10 highest rated Netflix series
top_10 = netflix_data.sort_values("rating",ascending=False).head(20)
top_10
```

```python
top_10["title"].value_counts()
```

```python
#Insight: Avatar: The Last Airbender and JoJo's Bizarre Adventure are the highest-rated series/anime on Netflix.

#Also, Avatar: The Last Airbender and JoJo's Bizarre Adventure are not Netflix original
```

```python
top_10["genre"].value_counts()
```

Insight: Among the top 10 highest-rated Netflix series the most popular genres are:

Animation, Action, Adventure
Crime, Drama, Mystery

### <FONT COLOR='#800080'>Top 10 Highest rated Netflix series

```python
plt.style.use("bmh")

plt.figure(figsize=(8,8))

plt.barh(top_10["title"],top_10["rating"],color="#0AD7E4",)
```

### <FONT COLOR='DeepPink'>Most popular genres among the top 10 Netflix series

```python
top_10["genre"].value_counts().plot(kind="barh",color="#7B68EE",figsize=(8,8))
```

### <FONT COLOR='DarkOrange'>Univariate Analysis



```python
# Histogram of ratings
plt.hist(netflix_data['rating'], bins=10, edgecolor='k', color='#FF6347')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()

```

```python
# Bar plot of genres
genre_counts = netflix_data['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values, color='Crimson')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Genre Distribution')
plt.xticks(rotation=45)
plt.show()
```

### <font color = 'Lightsalmon'>Genre Analysis

```python
# Bar plot of average ratings by genre
genre_avg_ratings = netflix_data.groupby('genre')['rating'].mean().sort_values(ascending=False)
plt.bar(genre_avg_ratings.index, genre_avg_ratings.values, color='#FF00FF')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Rating by Genre')
plt.xticks(rotation=45)
plt.show()
```

 ###  <font color = 'Gold'>Rating Analysis

```python
# Box plot of ratings by genre
sns.boxplot(data=netflix_data, x='genre', y='rating', color='DeepSkyBlue')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.title('Rating Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

### <font color = 'Khaki'>Vote Analysis

```python

plt.barh(top_10["title"],top_10["rating"], color = 'GreenYellow')
plt.xlabel('Votes')
plt.ylabel('Title')
plt.title('Distribution of Votes')
plt.show()
```
